# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
````

````python
#Function that returns the derivative of a polynomial using sympy
def calculate_derivative(equation):
    x = sp.symbols('x')
    function = sp.sympify(equation)
    derivative = sp.diff(function, x)
    return derivative
````

````python
polynomial = '3*x**4 - 16*x**3 + 18*x**2'
derivative = calculate_derivative(polynomial)
print(f"The derivative of {polynomial} is: {derivative}")
````

# Task 4: Create separate array of Minimas and maximas

````python
# Define the variable
x = sp.symbols('x')

# Define the polynomial function
polynomial = 3*x**4 - 16*x**3 + 18*x**2

# Compute the first derivative
first_derivative = sp.diff(polynomial, x)

# Find critical points
critical_points = sp.solve(first_derivative, x)

# Compute the second derivative
second_derivative = sp.diff(first_derivative, x)

# Classify critical points into maxima and minima arrays
arr_maximas = []
arr_minimas = []
for point in critical_points:
    # Evaluate the second derivative at the critical point
    second_derivative_at_point = second_derivative.subs(x, point)
    if second_derivative_at_point > 0:
        arr_minimas.append(point)
    elif second_derivative_at_point < 0:
        arr_maximas.append(point)

# Output the results
print("Polynomial:", polynomial)
print("First Derivative:", first_derivative)
print("Second Derivative:", second_derivative)
print("Critical Points:", critical_points)

print(f'the array of minimas is {arr_minimas} and the array of maximas is {arr_maximas}')
````

## **Task 5:** Compute Global Minima and Maxima

````python
# Evaluate the polynomial at the critical points
critical_values = [polynomial.subs(x, cp) for cp in critical_points]

# Find the global maximum and minimum
global_maximum = max(critical_values)
global_minimum = min(critical_values)

# Print the critical points and global extrema
print("Critical Points:", critical_points)
print("Global Maximum:", global_maximum)
print("Global Minimum:", global_minimum)
````

# **Task 6:** Visualize the Global Minima and Maxima:

````python
# Generate x values for plotting
x_vals = np.linspace(-2, 3, 500)

# Generate y values for the polynomial
y_vals = [polynomial.subs(x, val) for val in x_vals]

# Plot the polynomial
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='3*x**4 - 16*x**3 + 18*x**2', color='blue')

# Highlight the global maxima and minima
plt.scatter([cp.evalf() for cp in critical_points], [pv.evalf() for pv in critical_values], color='red', label='Extrema')
plt.scatter([critical_points[critical_values.index(global_maximum)].evalf()], [global_maximum.evalf()], color='green', label='Global Maximum')
plt.scatter([critical_points[critical_values.index(global_minimum)].evalf()], [global_minimum.evalf()], color='purple', label='Global Minimum')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the Polynomial 3*x**4 - 16*x**3 + 18*x**2 with Extrema')
plt.legend()
plt.grid(True)

plt.show()
````
